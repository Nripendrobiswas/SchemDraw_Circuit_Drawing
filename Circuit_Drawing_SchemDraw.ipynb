```python
'''
The following python script is designed for a BJT Miscellaneous Bias Configuration
where Collector feedback is taken while emitter resistance is considered 0.

                Collector Feedback Of BJT

        Fig:4.53
        Page:189
        Book:Electronic Devices and Circuit Theory
            By Robert L. Boylestad, Louis Nashelsky
            11th Edition

        Script Written By:Shirajul Abedin Shimul
        Registration No.2020338001
'''

##We first include our necessary modules for this script
##We are using schemdraw for our assigned task

import schemdraw
import schemdraw.elements as elm      ##Assigning the elements of schemdraw as elm to be called upon

##We assign d as our Drawing canvas for schemdraw elements
d=schemdraw.Drawing()

##Started by creating a BJT npn transistor :
##Made sure that the parameters of our npn BJT is described(ex:amplification factor)
d.config(unit=3)
transistor=elm.BjtNpn(circle=True,fill='None').label(r'$\beta$=120',color='red').fill("#63A7E8")
d+=transistor       ##We added our BJT to the drawing


##Created the 1st resistor or the Collector resistor:
##Made sure that the parameters of our 1st resistor is described

res1=elm.Resistor(at=transistor.collector).up().label('4.7K$\Omega$',loc='bottom',color='red').label('$R_{C}$',loc='top',color='red')
d+=res1            ##We added our resistor to the drawing


##Added a juction point for connection of elements:

dot1=elm.Dot(radius=.02,close=True,at=res1.start)
d+=dot1


##Defined the endpoint of our 1st resistor:
##We labeled endpoint as our Voltage source Vcc=20V connection point

dot2=elm.Dot(radius=.02,close=True,at=res1.end).label('$V_{cc}$=20V',color='red')
d+=dot2


##Created the 2nd resistor or the Base resistor:
##Made sure that the parameters of our 2nd resistor is described

res2=elm.Resistor(at=transistor.collector).left().label('680K$\Omega$',loc='bottom',color='red').label('$R_{B}$',loc='top',color='red')
d+=res2            ##We added our resistor to the drawing


##Created wire for the connections of elements:

l1=elm.Line(at=res2.end).down().length(.7)
d+=l1              ##Added 1st wire to the drawing


l2=elm.Line(at=transistor.base).to(l1.end)    ##Connected transistor base to base resistor through l2 wire
d+=l2              ##Added 2nd wire to the drawing

##Added another junction point:

dot3=elm.Dot(radius=.02,close=True,at=l2.end)
d+=dot3


##Added our 1st capacitor to the BJT base resistor and base junction point:
##We described the parameters for the capacitor

capacitor1=elm.Capacitor2(at=dot3.start).label('10$\mu$F',loc='top',color='red').label('$C_{1}$',loc='bottom',color='red')
d+=capacitor1      ##Added capacitor to the drawing


##Defined the endpoint of our 1st capacitor & its parameters:
##We labeled endpoint as our input Sinusoidal Voltage source (Vi) connection point

dot4=elm.Dot(radius=.02,close=True,at=capacitor1.end).label('$V_{i}$',loc='left',color='red')
d+=dot4


##Defined our 2nd capacitor & its parameters:
##We added it to the 1st junction point or collector terminal of BJT

capacitor2=elm.Capacitor2(at=dot1.start).right().label('10$\mu$F',loc='top',color='red').label('$C_{2}$',loc='bottom',color='red')
d+=capacitor2      ##Added capacitor to the drawing


##Deined endpoint of our 2nd capacitor:
##We labeled endpoint as our output Voltage terminal (Vo) connection point

dot5=elm.Dot(radius=.02,close=True,at=capacitor2.end).label('$V_{0}$',loc='right',color='red')
d+=dot5


##Added Ground to the circuit through emitter of our BJT

Ground=elm.Ground(at=transistor.emitter)
d+=Ground


d.save('Fig4.53_ElecDevice_RBoylstadPg189_11E.pdf') ###Saving Our Project

##Display our drawing

d.draw()
```

```python
#############################################################################################
#############################################################################################
### This python script will produce a voltage-divider biasing using Jfet                  ###
### Fig.7.21                                                                              ###
### Page: 434                                                                             ###
### Book: Electronic Devices and Circuit Theory                                           ###
### By Robert L. Boylestad, Louis Nashelsky                                               ###
### 11th Edition                                                                          ###
#############################################################################################
### Student_Registration_number: 2020338025                                               ###
#############################################################################################
#############################################################################################



# Using the python package "SchemDraw"
# For more about the SchemDraw package: https://schemdraw.readthedocs.io/en/latest/index.html

# The greek leter is produced with Unicode
# CTRL+Shift+u  then type 03b2 for Beta, 03a9 for Omega, 00b5 for Micro and then press enter key
#for needed libary funcation for draw a voltage-divider biasing using Jfet
import schemdraw
import matplotlib
from schemdraw import elements as elm
with schemdraw.Drawing() as d:
    # basic configuration of drawing
    d.config(unit=3)
    #N-type JFet and teminal also add  I_DSS and V_p with value
    jfetn=elm.JFetP(circle=True, fill='skyblue').reverse().label('$I_{DSS} = 8 mA$\n$V_P = -4 V$',loc='right')
    d+=jfetn
    #for drain resistor R_D=2.4kΩ
    d+=elm.Resistor().at(jfetn.drain).up().label('$R_D$', loc='top').label('$2.4kΩ$',loc='bottom')
    #establish an anchor point at the termination of the drain resistor
    d.push()
    #draw a line originating
    d+=elm.Line().tox(jfetn.drain[0] + 2)
    #for +16V V_DD
    d+=elm.Dot(open =True).label('$+16V$',loc='right')
    #return to the previously mentioned anchor point
    d.pop()
    d+=elm.Line().left()
    #for resistor R_1=2.1MΩ
    d+=elm.Resistor().down().label('$R_1$',loc='top').label('$2.1MΩ$',loc='bottom')
    #draw a line originating
    d+=elm.Line().toy(jfetn.gate)
    #attach an anchor to the end of the initial resistor
    d.push()
    #draw a line originating
    d+=elm.Line().tox(jfetn.gate)
    #denotes gate & darin & source terminal
    #for gate
    d += elm.Dot().at((jfetn.gate[0] - .5, jfetn.gate[1])).label('G', ofst=(0, .1))
    #for darin
    d += elm.Dot().at((jfetn.drain[0], jfetn.drain[1] + .5)).label('D', ofst=(.2, 0))
    #for source
    d += elm.Dot().at((jfetn.source[0], jfetn.source[1] - .5)).label('S', ofst=(.2, -.5))
    # Move the anchor point backward
    d.pop()
    #Place another anchor point at the end of the first resistor
    d.push()
    #for capacitor(input)
    d+=elm.Capacitor2().left().label('$C_1$',loc='top').label('$5 μF$',loc='bottom')
    #for input voltage
    d+=elm.Dot(open=True).label('$V_i$',loc='left')
    #Return to the previously mentioned anchor point
    d.pop()
    d+=elm.Dot()
    #for resistor R_2=270kΩ
    d+=(r:= elm.Resistor().down().label('$R_2$',loc='top').toy(-4.5).label('$270kΩ$',loc='bottom'))
    #ground the source resistor
    d+=elm.Resistor().at(jfetn.source).down().label('$R_S$', loc='top').toy(-4.5).label('$1.5kΩ$',loc='bottom')
    d+=elm.Ground()
    #connect the output terminal to a capacitor
    d+=elm.Capacitor2().at((jfetn.drain[0], jfetn.drain[1] + .5)).right().label('$C_2$', loc='bottom').label('$10 μF$',loc='top')
    #for V_0
    d+=elm.Dot(open=True).label('$V_O$',loc='right')
    #draw a line originating from the source point
    d+=elm.Line().at((jfetn.source[0], jfetn.source[1] - .5)).right().tox(jfetn.source[0] + 2.2)
    #connect a capacitor in parallel with the source resistor
    d+= elm.Capacitor2().down().label('$C_S$',loc='top').toy(-4.5).label('$20 μF$',loc='bottom')
    #ground the capacitor
    d+=elm.Line().tox(r.end)
    #for save figure as pdf
    d.save('Fig7.21_ElecDevice_RBoylstadPg434_11E.pdf')
    #to draw
    d.draw()
```

```python
# Import the schemdraw library and elements module
import schemdraw
import schemdraw.elements as elm

# Create a schematic drawing context
with schemdraw.Drawing() as d:
    # Add a P-channel junction field-effect transistor (JFET) with specific attributes and labels
    d += ( J:= elm.JFetP(circle=True).fill('skyblue').reverse().label('$I_{DSS}$ \n $V_{P}$','rgt'))

    # Add a line from the gate of the JFET to the left
    d += elm.Line(at=J.gate).length(0.5).left()

    # Add a dot labeled 'G'
    d += elm.Dot().label('G')

    # Add a line with a length of 1
    d += elm.Line().length(1)

    # Add a line with a length of 2 going down
    d += elm.Line().length(2).down()

    # Add a ground symbol
    d += elm.Ground()

    # Add a line from the source of the JFET going down
    d += elm.Line(at=J.source).length(0.5).down()

    # Add a dot labeled 'S' on the left
    d += (D1:= elm.Dot().label('S','lft'))

    # Add a capacitor to the right of 'D1' with a label '$C_{1}$' at the bottom
    d += elm.Capacitor2(at=D1.end).right().label('$C_{1}$','bot')

    # Add an open dot labeled '$V_{i}$' on the right
    d += elm.Dot(open=True).label('$V_{i}$','rgt')

    # Add a resistor to the right of 'D1' with a label '$R_{S}$' at the bottom, going down
    d += elm.Resistor(at=D1.end).length(2.5).label('$R_{S}$','bot').down()

    # Add an open dot labeled '$V_{SS}$' on the right
    d += elm.Dot(open=True).label('$V_{SS}$','rgt')

    # Add a line from the drain of the JFET going up
    d += elm.Line(at=J.drain).length(0.5).up()

    # Add a dot labeled 'D' on the left
    d += (D2:= elm.Dot().label('D','lft'))

    # Add a capacitor to the right with a label '$C_{2}$'
    d += elm.Capacitor2('right').label('$C_{2}$')

    # Add an open dot labeled '$V_{o}$' on the right
    d += elm.Dot(open=True).label('$V_{o}$','rgt')

    # Add a resistor to the right of 'D2' with a label '$R_{D}$' at the bottom, going up
    d += elm.Resistor(at=D2.end).length(4).label('$R_{D}$','bot').up()

    # Add an open dot labeled '$V_{DD}$' on the right
    d += elm.Dot(open=True).label('$V_{DD}$','rgt')

    # Add a current label '$I_{D}$' at a specific position with color 'skyblue'
    d += elm.CurrentLabel().label('$I_{D}$','bot').reverse().scale(0.3).at([0.5,3.5]).color('skyblue')

    # Draw the schematic
    d.draw()
```

```python
"""

    This python script will generate a JFET biasing configuration
    Fig.7.60
    Page: 459
    Book: Electronic Devices and Circuit Theory By Robert L. Boylestad,(11th Edition)

    Student Name  : Nahidul Hassan
    Student Reg no: 2020338036

"""
## first Import the requird packages
import schemdraw
import schemdraw.elements as elm

## creating a drawing object denoted by d
with schemdraw.Drawing() as d :

    ## Basic Configuration
    d.config(unit=3)

    ### Started with N-type jfet
    jfet = elm.JFetP(circle=True,fill='skyblue').reverse()
    d += jfet

    ### JFET's terminals
    d += elm.Dot().at((jfet.drain[0],jfet.drain[1]+0.5)).label('D',loc='left')#drain terminal
    d += elm.Dot().at((jfet.gate[0]-0.5,jfet.gate[1])).label('G',loc='top').label("$+$",loc='bot',color='skyblue',ofst=(0.25,0)).label("$V_{{GS}_Q}$",loc='bot',color='skyblue',ofst=(.55,-0.4))#gate terminal
    d += elm.Dot().at((jfet.source[0],jfet.source[1]-0.5)).label('S',loc='right',ofst=(0,-0.35)).label("$-$",loc='top',color='skyblue',ofst=(-0.35,-0.25))#source terminal

    ### Drain terminal
    d += elm.Line().at(jfet.drain).up().toy(((jfet.drain[0],jfet.drain[1]+0.5)))
    d += (q:= elm.Resistor().at(((jfet.drain[0],jfet.drain[1]+0.5))).up().label('$R_D$',loc='top').label('3.9K$\Omega$',loc='bottom'))#Drain reisitance
    d += (w:= elm.Dot(open='true').label('16V',loc='right'))## Biasing Voltage of 16V
    d += elm.CurrentLabel(reverse='true',length=0.75,color='skyblue').at(0.85*(q.end)).flip().label('$I_{D_Q}$',color='skyblue',loc='bot')

    ## Gate terminal
    d += elm.Line().at(jfet.gate).left().tox(jfet.gate[0]+1)
    d.push()#creating a anchor

    ## Input end with coupling capacitor
    d += elm.Capacitor2().left().label('$C_1$',loc='top').label('0.05$\mu$F',loc='bottom')#input coupling capacitor
    d += elm.Dot(open='true').label('$V_i$',color='skyblue',loc='left')#input voltage terminal
    d.pop()#back to anchor point

    ## Gate Resistor
    d += elm.Resistor().label('$R_G$',loc='top').down().tox(jfet.gate[0]+3).label('1M$\Omega$',loc='bottom')#
    d += elm.Line().right().tox(jfet.source[0])

    ## Source terminal
    d += elm.Line().at(jfet.source).down().toy((jfet.source[0],jfet.source[1]-0.75))
    d += elm.Resistor().at((jfet.source[0],jfet.source[1]-0.75)).down().label('$R_S$',loc='top').label('1.6K$\Omega$',loc='bottom')#source resistance
    d += elm.Ground()#source grounded
    d += elm.Line().at((jfet.source[0],jfet.source[1]-0.5)).right().tox(jfet.source[0]+2)

    ## Bypass Capacitor
    d += elm.Capacitor2().down().tox(jfet.source[0]+5.09).label('40$\mu$F',loc='bottom') #bypass capacitor
    d += elm.Line().left().tox(jfet.source[0]) #bypass capacitor grounded

    ## Output end with coupling capacitor
    d += elm.Capacitor2().at((jfet.drain[0],jfet.drain[1]+0.5)).right().label('$C_2$',loc='bottom').label('0.05$\mu$F',loc='top')#Output coupling capacitor
    d += elm.Dot(open='true').label('$V_o$',color='skyblue',loc='right')#output voltage terminal

    ## Save and Display
    d.save('Fig7.60_ElecDevice_RBoylstadPg459_11E.pdf') #saves the drawing
    d.draw() # Displays the drawing in the window

```

```python
        ###############################################################################
        ###############################################################################
        ###                     This python script will produce                     ###
        ###                    A Vlotage Devider Cascaded Circuit                   ###
        ###                                  using                                  ###
        ###                             a BJT and a JFET                            ###
        ###                                                                         ###
        ###                     Fig_7.96                                            ###
        ###                     Page: 478                                           ###
        ###                     Book: Electronic Devices and Circuit Theory         ###
        ###                             By Robert L. Boylestad, Louis Nashelsky     ###
        ###                             11th Edition                                ###
        ###                     Code Written by: Md Ishraqure Rahman                ###
        ###                                      Dolon                              ###
        ###                     Registration:    2020338011                         ###
        ###############################################################################
        ###############################################################################


# Using the python package "SchemDraw"
# For more about the SchemDraw package please visit
# Project Link: https://schemdraw.readthedocs.io/en/latest/index.html

# The greek leter is produced with Unicode

#import necessery moduels
import schemdraw
import schemdraw.elements as elm

####    Start to set the circuit with schemdraw.Drawing()   ####
with schemdraw.Drawing() as circuit:

    circuit.config(unit=3)
    ####    Beginig of the Upper Half BJT loop  ####

        ####    Inserting The BJT Transistor as T1  ####
    circuit+=(T1:= elm.BjtNpn(circle=True).label(r"$\beta$ =160",color="red").fill("#AAFA37"))
    circuit+=(L1:= elm.Line().up().at(T1.collector).length(.5))
    circuit+=(D1:= elm.Dot(radius=.05).at(L1.end))

            ####    Seting the indicator for Vc ####
    circuit+=(L2:= elm.Line().right().length(1))
    circuit+=(D2:= elm.Dot(radius=.05,open=True).at(L2.end)
              .label("$V_{c}$",loc="right",color="red"))

        ####    Inserting The first Resistor as R1  ####
    circuit+=(R1:= elm.Resistor("up",at=(T1,"collector"))
              .label("$1.1K\Omega$",ofst=-1.75,color="#03AFB8"))
    circuit+=(D3:= elm.Dot(radius=.05).at(R1.end))

    circuit.push()  ####    Remembers the position for furture usage    ####

            ####    Seting the indicator for 20V Vcc supply ####
    circuit+=(L3:= elm.Line().right().length(1))
    circuit+=(D4:= elm.Dot(radius=.05,open=True).at(L3.end)
              .label("$20V$",loc="right",color="red"))

    circuit.pop()   ####    Retarning to previous position and continue Drawing ####

            ####    Simple lines and Dots
    circuit+=(L4:= elm.Line().left().length(2))
    circuit+=(D5:= elm.Dot(radius=.05).at(L4.end))

    circuit.push()  ####    Remembers the position of the start of 2nd Resistor    ####

        ####    Inserting the 2nd Resistor as R2 ####
    circuit+=(R2:= elm.Resistor("down").label("$330K\Omega$",ofst=-1.75,color="#03AFB8"))
    circuit+=(L5:= elm.Line().down().length(.7))
    circuit+=(elm.CurrentLabel(ofst=.4,length=.75).at(T1.emitter)
              .down().label("$I_E$",ofst=-.75).color('#9364FE'))     ####    For current arrow ####
    circuit+=(L6:= elm.Line().endpoints(L5.end,T1.base))    ####    compliting the loop by
                                                         # Connecting with base of the BJT T1
    circuit+=(elm.CurrentLabel(ofst=-.4,length=.75).at(L6.center)
                .right().label("$I_B$",ofst=-.75).color('#9364FE'))  ####    For current arrow ####

    ####    End of Upper Half BJT loop  ####

    ####    Beginig of the Lower Half JFET loop  ####

        ####    Setting the indicator for Vd    ####
    circuit+=(L7:= elm.Line().down().length(.5).at(T1.emitter))
    circuit+=(D5:= elm.Dot(radius=.05).at(L7.end))
    circuit.push()  ####    Pointer for returning after seting the Vd   ####
    circuit+=(L8:= elm.Line().right().length(1))
    circuit+=(D6:= elm.Dot(radius=.05,open=True).at(L8.end)
              .label("$V_{D}$",loc="right",color="red"))

        ####    Setting the JFET as T2  ####
    circuit.pop()   ####    Begining of the 2nd Transistor
    circuit+=(L9:= elm.Line().down().length(.5))
    circuit+=(T2:= elm.JFetP(circle=True).anchor("source").at(L9.end)
              .left().label("$I_{DSS}=6mA$\n$V_{P}=-6V$",ofst=-3.75,color="red").fill("#37FAA1"))
    circuit+=(elm.CurrentLabel(ofst=.4,length=.75).at(T2.source)
              .down().label("$I_{D_{Q}}$",ofst=-.75)
              .color('#9364FE'))   ####    For the Current arrow    ####

        ####    Resistor below the 2nd Transistor   ####
    circuit+=(L10:= elm.Line().down().length(.5)
              .label("$V_{GS_{Q}}$  __",ofst=.3,color="red").at(T2.drain))
    circuit+=(R3:= elm.Resistor("down",at=L10.end)
              .label("$1.2K\Omega$",ofst=-1.75,color="#03AFB8"))
    circuit+=(G1:= elm.Ground().color("#E8A504"))    ####    Connenting to the Ground    ####

        ####    The resistor to the far left 4th Resistor as R4   ####
    circuit+=(L11:= elm.Line().left().at(L4.end).length(2.5))
    circuit+=(R4:= elm.Resistor("down",at=L11.end)
              .label("$91K\Omega$",ofst=-1.75,color="#03AFB8")) ####    Resistor label  ####
    circuit+=(L12:= elm.Line().down().at(R4.end).length(2.8))
    circuit+=(D7:= elm.Dot(radius=.05).at(L12.end)
              .label("$V_{G}$",loc="left"
                     ,color="red")) ####    For the Gate Voltage label (Vg) ####

        ####    Resistor under the 4th Resistor as R5   ####
    circuit+=(L13:= elm.Line().right().at(L12.end).length(3))
    circuit+=(L14:= elm.Line().endpoints(L13.end,T2.gate)
              .label("+",fontsize=20,loc="start"
                     ,ofst=-.4,color="red"))        ####    Line connects to the JFET Gate  ####
    circuit+=(R5:= elm.Resistor("down",at=L12.end)
              .label("$12K\Omega$",ofst=-1.75,color="#03AFB8")) ####    R5 Resistance   ####
    circuit+=(L15:= elm.Line().down().length(1.5))
    circuit+=(G2:= elm.Ground().at(L15.end).color("#E8A504"))    ####    Ground connection   ####
    ####    End of the Lower Half JFET loop  ####
    circuit.save("Fig7.96_ElecDevice_RBoylstadPg478_11E.pdf")
    circuit.draw()
```

```python
###############################################################################
###############################################################################
### This python script will produce a E-MOSFET drain-feedback configuration ###
### Fig.:8.40                                                               ###
### Page:509                                                                ###
### Book: Electronic Devices and Circuit Theory                             ###
### By Robert L. Boylestad, Louis Nashelsky                                 ###
### 11th Edition                                                            ###
###############################################################################
### Student_Registration_number: 2020338058                                 ###
###############################################################################
###############################################################################


# Using the python package "SchemDraw"
# For more about the SchemDraw package: https://schemdraw.readthedocs.io/en/latest/index.html

# The greek leter is produced with Unicode
# CTRL+Shift+u  then type 03b2 for Beta, 03a9 for Omega, 00b5 for Micro and then press enter key

#The schemdraw module allows for drawing circuit elements.
import schemdraw

#To make a circuit diagram, a schemdraw.Drawing and schemdraw.elements is created
import schemdraw.elements as elm    #assigning elements as elm
with schemdraw.Drawing() as d :     #assigning Drawing as d
    d.config(unit=3)
#At first I draw an Enhancement MOSFET and I labeled
    d += (T1 := elm.NMos(circle=True).fill('skyblue')
          .label('$  I_{D(on)}$=6 mA \n $V_{GS(on)}$= 8 V \n $V_{GS(Th)}$=3 V \n $g_{os}$=20 $\mu$S', 'bot',
                 ofst=(2, 1))
          )


#To create a wire from drain to up
    d += elm.Line('up', at=(T1, 'drain')).length(2)
#To create a Drain Resistor and labeled
    d += elm.Resistor().label('2 k$\Omega$', loc='bot').label('$R_{D}$').length(2.5)
#To labeled endpoint as Voltage source 12V connection point
    d += elm.Dot(open=True).up().label('12V','bot')
#To create an junction point I used a wire from drain to junction point then I used dot as junction point
    d += elm.Line('up', at=(T1, 'drain')).length(2)
    d += elm.Dot(open=False)
#To write a capacitor from junction to right side and I labeled
    d += elm.Capacitor2().right().label('1 $\mu F$', 'bot')
#To indicate Zo I used currentLabel
    d += elm.CurrentLabel(0.1, 2).left().scale(0.4).at([2.5, 1.5]).label('$Z_{o}$', loc='bottom').color('skyblue')
#To labeled end point as output voltage
    d += elm.Dot(open=True).label('$V_{o}$', loc='rgt',color='skyblue')


#To write a wire from gate to left
    d += elm.Line('left',at=(T1,'gate')).length(2)
#To indicate a Junction point I used Dot
#Also junction defined as D2
    d +=(D2:= elm.Dot(open=False))
#To write a Capacitor from Junction point to the left side and labeled
    d += elm.Capacitor2().label('1 $\mu F$', 'bot').length(3)
#To labled ebd point as Input voltage
    d +=elm.Dot(open=True).label('$V_{i}$','lft',color='skyblue')
#To indicate Zi I used current label
    d += elm.CurrentLabel(0.1, 2).right().scale(0.4).at([-4.2, -1.5]).label('$Z_{i}$', loc='bottom').color('skyblue')
#To write a Resistor from Junction D2 to up
    d += elm.Resistor(at=D2.start).length(2.5).up().label('10 M$\Omega$','bot').label('$R_{F}$')
    d += elm.Line().right()


#To write Ground which is start from Source
    d += elm.Ground(at=(T1,'source'))

#To save
    d.save('Fig8.40_ElecDevice_RBoylstadPg509_11E.pdf')
    #To draw
    d.draw()
```
